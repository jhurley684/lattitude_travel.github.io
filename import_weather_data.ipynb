```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load: ('resources/purchase_data.csv')

# Read Purchasing File and store into Pandas data frame
weather_data = pd.read_csv("instructions/resources/cities.csv")
weather_data_df = pd.DataFrame(weather_data)
```

```python
# Display data preview for easy reference

weather_data_df = weather_data_df[["City","Cloudiness","Country","Date","Humidity", "Lat","Lng","Max Temp", "Wind Speed", "Date - Time"]]
weather_data_df
```

```python
# dropping the 0th and the 1st index

# resetting the DataFrame index
# weather_data_df
weather_data_df.reset_index()
weather_data_df = weather_data_df.iloc[1: , :]
weather_data_df
```

```python
html = weather_data_df.to_html()
```

```python
# write html to file
text_file = open("weather_data.html", "w")
text_file.write(html)
text_file.close()
```
